#### Criando dataframe

```python
# Caminho da pasta onde estão os arquivos CSV
caminho_pasta = r"C:\Users\fluca\Desktop\Data Science\Projetos\projeto_tcc\imoveis_zapimoveis"

# Lista para armazenar os dataframes individuais
dataframes = []

# Iterar por todos os arquivos no diretório
for arquivo in os.listdir(caminho_pasta):
    if arquivo.endswith('.csv'):  # Verifica se o arquivo é um CSV
        caminho_arquivo = os.path.join(caminho_pasta, arquivo)  # Caminho completo do arquivo
        print(f"Lendo o arquivo: {caminho_arquivo}")
        
        # Lendo o CSV em um dataframe e adicionando à lista
        df = pd.read_csv(caminho_arquivo)
        dataframes.append(df)

# Concatenar todos os dataframes em um único
df_final = pd.concat(dataframes, ignore_index=True)

# Exibir as primeiras linhas do dataframe final
print(df_final.head())

# Opcional: Salvar o dataframe final em um novo arquivo CSV
# df_final.to_csv(r"C:\Users\fluca\Desktop\Data Science\Projetos\projeto_tcc\imoveis_zapimoveis\consolidado.csv", index=False)
```

#### Transformação

```python
# Verificar valores que não correspondem ao padrão esperado
padrao = r'^\d+\s+(quartos|quarto)$'  # Exemplo: '3 quartos', '1 quarto'
valores_incorretos = df[~df['Quartos'].str.match(padrao, na=False)]

# Exibir os valores incorretos
print("Valores incorretos encontrados:")
print(valores_incorretos['Quartos'].unique())

# Exibir as linhas com valores incorretos
print(valores_incorretos)
```

```python
# Contar registros com valores 'Nao encontrado' em cada coluna
contagem_incorretos = df.isin(['Nao encontrado']).sum()
print("Contagem de valores 'Nao encontrado' por coluna:")
print(contagem_incorretos)
```

```python
df.rename(
    columns={
        'ID': 'id',
        'Área': 'area',
        'Quartos': 'quartos',
        'Banheiros': 'banheiros',
        'Vagas de Carro': 'vagas_carro',
        'Preço': 'preco',
        'Endereço': 'endereco',
        'Itens Adicionais': 'itens_adc',
        'Códigos': 'cod',
        'Texto do Anúncio': 'texto',
    },
    inplace=True
)
```

```python
df.head()
```

```python
# Extrair apenas o código do anunciante da coluna 'cod'
df['cod'] = df['cod'].str.extract(r'Codigo do anunciante:\s*([\w\-]+)')

# Exibir as primeiras linhas para confirmar a extração
df.head()
```

```python
# Remover 'm2' e espaços da coluna 'area'
df['area'] = df['area'].str.replace(r'm2|\s+', '', regex=True)

# Remover 'quartos', 'quarto' e espaços da coluna 'quartos'
df['quartos'] = df['quartos'].str.replace(r'quartos?|[\s]', '', regex=True)

# Remover 'banheiros', 'banheiro' e espaços da coluna 'banheiros'
df['banheiros'] = df['banheiros'].str.replace(r'banheiros?|[\s]', '', regex=True)

# Remover 'R$', '.' e espaços da coluna 'preco'
df['preco'] = df['preco'].str.replace(r'R\$|\.|\s+', '', regex=True)

# Exibir as primeiras linhas para verificar as alterações
df.head()
```

```python
# Reordenar as colunas para que 'cod' seja a segunda
colunas = list(df.columns)  # Obter a lista atual de colunas
colunas.remove('cod')  # Remover 'cod' da lista atual
colunas.insert(1, 'cod')  # Inserir 'cod' na segunda posição

# Reorganizar o dataframe
df = df[colunas]

# Exibir as primeiras linhas para confirmar a reordenação
df.head()
```

```python
# Função para processar a coluna
def process_vagas(value):
    if "vaga" in value:  # Verifica se contém "vaga" ou "vagas"
        return None
    return value  # Retorna o valor total original

# Criar a nova coluna com base na lógica
df2["adc_1"] = df2["vagas_carro"].apply(process_vagas)

df2.head()
```

```python
# Função para processar a coluna
def process_vagas2(value):
    if "vaga" in value:  # Verifica se contém "vaga" ou "vagas"
        return value
    return None  # Substitui valores diferentes por None

# Aplicar a lógica diretamente na coluna 'vagas_carro'
df2["vagas_carro"] = df2["vagas_carro"].apply(process_vagas2)

df2.head()
```

```python
# Função para processar a coluna
def process_vagas3(value):
    if value is not None and "suite" in value:  # Verifica se contém "suite"
        return None  # Substitui por None
    return value  # Retorna o valor original

# Certifique-se de que 'adc_1' existe antes de aplicar
if "adc_1" in df2.columns:
    df2["adc_2"] = df2["adc_1"].apply(process_vagas3)
else:
    print("A coluna 'adc_1' não existe no DataFrame.")

# Exibir as primeiras linhas do DataFrame
df2.head()
```

```python
# Função para processar a coluna
def keep_only_suite(value):
    if value is not None and "suite" in value:  # Verifica se contém "suite"
        return value  # Mantém o valor original
    return None  # Substitui os demais por None

# Aplicar a lógica diretamente na coluna 'adc_1'
df2["adc_1"] = df2["adc_1"].apply(keep_only_suite)

# Exibir o DataFrame atualizado
df2.head()
```

```python
# Separando as informações com base nas vírgulas (limitando o número de colunas)
separado = df2['itens_adc'].str.split(', ', expand=True)

# Garantindo que sempre haja pelo menos 6 colunas
num_cols = separado.shape[1]  # Número de colunas geradas
colunas = ['Área', 'Quartos', 'Banheiros', 'Vagas', 'Outros', 'Extra']
for i in range(num_cols, len(colunas)):
    separado[i] = None  # Adiciona colunas vazias se necessário

# Ajustando os nomes das colunas
separado.columns = colunas[:num_cols]

# Concatenando o DataFrame original com as novas colunas
df2 = pd.concat([df2, separado], axis=1)

# Resultado
df2.head()
```

### Geolocalização

```python
df7 = df6
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError
import time
```

```python
# Configura o geolocalizador
geolocator = Nominatim(user_agent="my_geocoder_application")

# Função para buscar latitude e longitude com tratamento de exceções e atraso
def get_lat_long(address):
    try:
        location = geolocator.geocode(address, timeout=15)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

# Adiciona um intervalo entre requisições
latitudes = []
longitudes = []

for address in df7['endereco']:
    lat, lon = get_lat_long(address)
    latitudes.append(lat)
    longitudes.append(lon)
    time.sleep(3)  # Intervalo de 1 segundo

# Adiciona os resultados ao DataFrame
df7['Latitude'] = latitudes
df7['Longitude'] = longitudes

df7.head()
```
